## Analysis of the rewards

```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Paths
DATA_PATH = './data'
```

```python
# Define function for plotting mean reward
def plot_reward(rewards, ax, font_size=16, tick_size=12, alpha=0.7):
    # Compute mean reward
    x = np.arange(rewards.shape[0]) + 1
    y = np.mean(rewards, axis=1)
    # Set title and labels
    ax.set_title('Mean reward per episode', fontsize=font_size)
    ax.set_xlabel('Episode', fontsize=font_size)
    ax.set_ylabel('Reward (mean)', fontsize=font_size)
    # Set ticks size
    ax.tick_params(axis='both', which='major', labelsize=tick_size)
    # Make plot
    ax.plot(x, y, '.-', alpha=alpha)
    # Enable grid
    ax.grid()
    # Set boundaries
    ax.set_ylim(-1, 1)
```

### No obstacles

```python
# Load rewards
no_soft_no_sarsa = np.genfromtxt(DATA_PATH + '/no_obstacles/no_soft_no_sarsa/rewards.csv', delimiter=',')
no_soft_sarsa = np.genfromtxt(DATA_PATH + '/no_obstacles/no_soft_sarsa/rewards.csv', delimiter=',')
soft_sarsa = np.genfromtxt(DATA_PATH + '/no_obstacles/soft_sarsa/rewards.csv', delimiter=',')

# Initialize plot
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
# Make scatter
plot_reward(no_soft_no_sarsa, ax=ax)
# Make scatter
plot_reward(no_soft_sarsa, ax=ax)
# Make scatter
plot_reward(soft_sarsa, ax=ax)
# Show lables
ax.legend(['No softmax, Q-learning', 'No softmax, SARSA', 'Softmax, SARSA'], loc=4, fontsize=12)
# Save figure
plt.savefig(DATA_PATH + '/no_obstacles.png', dpi=100)
# Show plot
plt.show()
```
